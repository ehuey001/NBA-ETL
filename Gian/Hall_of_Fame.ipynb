```python
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
from bs4 import BeautifulSoup

import time
import pandas as pd

!which chromedriver
```

```python
url1 = "https://en.wikipedia.org/wiki/List_of_players_in_the_Naismith_Memorial_Basketball_Hall_of_Fame"
```

```python
tables = pd.read_html(url1)
hall = tables[0]
hall.head()
```

```python
list_hall = hall.loc[hall["Year"]>1987, :]
list_of_hof = list_hall["Inductees"].to_list()
list_of_hof
```

```python
halloffame_df = pd.DataFrame(list_of_hof)
halloffame_df.head()
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.basketball-reference.com/'
browser.visit(url)
```

```python
player_totals = pd.DataFrame()

for player in list_of_hof:
    try:
        html = browser.html
        soup = BeautifulSoup(html, 'html.parser')
        browser.find_by_css('.ac-input').first.fill(player)
        browser.find_by_value('Search').first.click()
    
        time.sleep(2)
        
        table = soup.find('div', {'id':'div_totals'})
        df = pd.read_html(str(table))[0]
        df['Player'] = player
    
        player_totals = player_totals.append(df, ignore_index = True)
    except:
        pass
```

```python
player_totals
```

```python
player_totals.to_csv('Outputs/HOF_stats.csv', encoding='utf-8')
```
